```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the Impact_of_Remote_Work_on_Mental_Health.csv.
import pandas as pd
mental_df = pd.read_csv("Impact_of_Remote_Work_on_Mental_Health.csv", na_values=[''], keep_default_na=False)
mental_df.head()
```

```python
mental_df.info()
```

```python
print(mental_df.isna().sum())  # Column-wise count of NaN values
print(mental_df.isna().any())  # Check if any NaN exists in each column
print(mental_df.isna().sum().sum())  # Total count of NaN in the dataset
```

```python
# Recode the target variable for binary classification
mental_df['Mental_Health_Condition'] = mental_df['Mental_Health_Condition'].replace({
    'Anxiety': 'Yes',
    'Depression': 'Yes',
    'Burnout': 'Yes',
    'None': 'No'
})
mental_df.head()
```

```python
# Count occurrences of 'Yes' and 'No' in the Mental_Health_Condition column
counts = mental_df['Mental_Health_Condition'].value_counts()

# Display the counts
print(counts)
```

```python
# Drop the non-beneficial ID columns, 'Employee ID' and 'Age'.
mental_df = mental_df.drop(columns = ['Employee_ID', 'Age'])
mental_df
```

```python
mental_df.nunique()
```

```python
y = mental_df['Mental_Health_Condition']
X = mental_df.drop(columns='Mental_Health_Condition')
```

```python
y[:5]
```

```python
X = pd.get_dummies(X)
# Split the dataset using train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=16)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Import the KNeighborsClassifier module from sklearn
from sklearn.neighbors import KNeighborsClassifier

# Instantiate the KNeighborsClassifier model with n_neighbors = 3
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# Train the model using the training data
knn.fit(X_train_scaled, y_train)
```

```python
# Create predictions using the testing data
y_pred = knn.predict(X_test_scaled)
```

```python
from sklearn.metrics import classification_report
# Print the classification report comparing the testing data to the model predictions
print(classification_report(y_test, y_pred))
```
